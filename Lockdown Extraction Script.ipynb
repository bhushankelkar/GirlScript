```python
import twint
import nest_asyncio
import pandas as pd
```

```python
dates = ['2020-03-25',
         '2020-03-26',
         '2020-03-27',
         '2020-03-28',
         '2020-03-29',
         '2020-03-30',
         '2020-03-31',
         '2020-04-01',
         '2020-04-02',
         '2020-04-03',
         '2020-04-04',
         '2020-04-05',
         '2020-04-06',
         '2020-04-07',
         '2020-04-08',
         '2020-04-09',
         '2020-04-10',
         '2020-04-11',
         '2020-04-12',
         '2020-04-13',
         '2020-04-14',
         '2020-04-15',
         '2020-04-16',
         '2020-04-17',
         '2020-04-18',
        ]
```

```python
df = pd.DataFrame()
for i in range(0,len(dates)-1):
    nest_asyncio.apply()
    c = twint.Config()
    c.Search = "(corona OR coronavirus OR covid OR pandemic OR lockdown OR quarantine OR covid-19 OR ppe)" 
    c.Pandas = True
    c.Pandas_clean = True
    c.Limit = 1000
    #c.Count = 1000
    c.Geo = "28.5934,77.2223,2000km"# Run
    c.Since = dates[i]
    c.Until = dates[i+1]
    c.Lang = 'en'
    twint.run.Search(c)
    df_temp = pd.DataFrame(twint.storage.panda.Tweets_df)
    df = df.append(df_temp , ignore_index = True)
    print("\n Current: ",dates[i], " ", df.shape)
```
